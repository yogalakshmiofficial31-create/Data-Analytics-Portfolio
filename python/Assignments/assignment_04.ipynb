PART-A - functions
1.Calculate Total sales

```python
sales = [12000, 15000, 18000, 9000, 21000, 17000]
```

```python
def calculate_total_sales(sales):
    return sum(sales)
print('Total sales:',calculate_total_sales(sales))
```

2.calculate Average Sales

```python
def calculate_average_sales(sales):
    return sum(sales)/len(sales)
print("Average_sales:",calculate_average_sales(sales))
```

3.Highest sale

```python
def highest_sale(sales):
    return max(sales)
print('Highest sale:',highest_sale(sales))
```

4.Lowest sale

```python
def lowest_sale(sales):
    return min(sales)
print('Lowest sale:',lowest_sale(sales))
```

Part - B
Functions Arguments & Return Values

```python
def sales_growth(current_month,previous_month):
    growth = ((current_month - previous_month) / previous_month) * 100
    return growth
result = sales_growth(25000,20000)
print('sales growth',result,'%')
```

Part- C - Lambda functions
1.Calculate Discounted price

```python
discount = lambda amount:amount * 0.9
price = 10000
print('original price',price)
print('discount price',discount(price))
```

2. Convert Customer Names to Uppercase

```python
customers = ["john", "mary", "david"]
customers_upper = [name.upper() for name in customers]
print(customers_upper)
```

3.sort products by price

```python
products = [("laptop",60000),
            ("mouse",500),
            ("keyboard",1500),
            ("moniter",12000)
]
sorted_products = sorted(products,key=lambda product: product[1])
print(sorted_products)
```

Part D - Recursion

```python
def factorial(n):
    if n==1:
        return 1
    return n * factorial(n-1)
print("factorial of 5:",factorial(5))
```

Additional Task

```python
def countdown(n):
    if n==0:
        return
    print(n)
    countdown(n-1)
countdown(10)
```
