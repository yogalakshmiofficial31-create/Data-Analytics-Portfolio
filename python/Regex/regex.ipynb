```python
import re
text = '1,2,3,4,5,6,7'
data = re.findall(r'\d',text)
print(data)
```

```python
#\d -Digits
#\d+ add numbers
#[a-z] -> a-z(lowercase)
#[A-Z] -> A-Z (Uppercase)
#\s spaces
#\w wordcharacters (letter,numbers,underscores)
```

```python
#Quantifiers
#\d+ one or more example['123']]
# * ,{10}-10 digits
```
